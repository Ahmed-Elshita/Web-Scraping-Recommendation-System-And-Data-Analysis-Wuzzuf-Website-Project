```python
# import libraries 

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##Step 1: Read CSV File
df = pd.read_csv("Wuzzuf WebScraper Dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.fillna('not found',inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isna().sum()
```

```python
df.drop(columns=['date','dateNum'],inplace =True)
```

```python
df.describe()
```

```python
company_name = df['company_name'].value_counts()
company_name = pd.DataFrame(company_name)
company_name
```

```python
f , ax = plt.subplots(figsize=(6,15))

sns.barplot(y=company_name.index , x= company_name.company_name ,data = company_name )
plt.title('Number of job offers in the company',size =25)
plt.xlabel('Number of job offers')
plt.ylabel('company')
plt.show()
```

```python
jop_type = df['jop_type'].value_counts()
jop_type = pd.DataFrame(jop_type)
jop_type
```

```python
plt.pie(jop_type, labels=jop_type.index , explode = [0,0.6],autopct = '%1.1f%%')
plt.title('jop type',size =25)
plt.show()
```

```python
jop_tital = df['jop_tital'].value_counts()
jop_tital = pd.DataFrame(jop_tital)
jop_tital
```

```python
f , ax = plt.subplots(figsize=(6,15))

sns.barplot(y=jop_tital.index , x= jop_tital.jop_tital ,data = jop_tital )
plt.title('Number of job title ',size =25)
plt.xlabel('Number of job title')
plt.ylabel('job title')
plt.show()
```

```python
jop_level = df['jop_level'].value_counts()
jop_level = pd.DataFrame(jop_level)
jop_level
```

```python
plt.pie(jop_level, labels=jop_level.index , explode = [0,0.1,0.2],autopct = '%1.1f%%')
plt.title('jop level',size =25)
plt.show()
```

```python
exp = pd.DataFrame()
exp['min_exp'] = df['min_exp']
exp['max_exp'] = df['max_exp']
exp.head()
```

```python
exp.hist()
```

```python
min_exp = exp['min_exp'].value_counts()
min_exp = pd.DataFrame(min_exp)
min_exp
```

```python
sns.barplot(x= min_exp.index , y= min_exp.min_exp ,data = min_exp )
plt.title('min_exp',size =25)
plt.xlabel('min_exp')
plt.ylabel('count')
plt.show()
```

```python
max_exp = exp['max_exp'].value_counts()
max_exp = pd.DataFrame(max_exp)
max_exp
```

```python
sns.barplot(x= max_exp.index , y= max_exp.max_exp ,data = max_exp )
plt.title('max_exp',size =25)
plt.xlabel('max_exp')
plt.ylabel('count')
plt.show()
```
