# create excel file

```python

header = ['company_name','company_location','date','dateNum','jop_tital','jop_type','jop_level','min_exp','max_exp','related_jop_titel','strongKeyWord','specificWorde','description']

with open('Wuzzuf WebScraper Dataset.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    
```

# collect data from all container

```python
def scraping(containers):
    
    i=0
    for container in containers:
        try:
            jop_tital = container.findAll('a',{'class':"css-o171kl"})[0].text.strip()

            company_name = container.findAll('a',{'class':"css-17s97q8"})[0].text.replace ("-","").strip()

            company_location = container.findAll('span',{'class':"css-5wys0k"})[0].text.strip()

            jop_type = container.findAll('span',{'class':"css-1ve4b75 eoyjyou0"})[0].text.strip()

            jop_level = container.findAll('a',{'class':"css-o171kl"})[1].text.strip()

            if i<4:
                date = container.findAll('div',{'class':"css-4c4ojb"})[0].text.strip()
                dateNum = re.findall('[0-9]+',date)[0]

            else :
                date = container.findAll('div',{'class':"css-do6t5g"})[0].text.strip()
                dateNum = re.findall('[0-9]+',date)[0]

            related_jop_titel = ''
            related_jop_titels = container.findAll('a',{'class':"css-o171kl"})
            for k in related_jop_titels[2:]:
                related_jop_titel +=k.text.replace ("·","").replace ("/","").strip() + ' '
            related_jop_titel = related_jop_titel.strip()     

            strongKeyWord = ''
            strongKey = container.findAll('strong',{'class':"highlight"})
            for k in strongKey:
                strongKeyWord +=k.text.replace ("·","").replace ("/","").strip() + ' '
            strongKeyWord = strongKeyWord.strip()     


            description = ''
            descriptions = container.findAll('a',{'class':"css-5x9pm1"})
            for k in descriptions:
                description +=k.text.replace ("·","").replace ("/","").strip() + ' '
            description = description.strip()     

            exp = container.findAll('span',{'class':""})[0].text.strip()
            exp = re.findall('[0-9]+',exp)
            min_exp = exp[0]
            max_exp = exp[-1]


            specificWorde = ''
            specificWordes = container.findAll('span',{'class':""})
            for k in specificWordes:
                specificWorde +=k.text.replace ("·","").replace ("/","").strip() + ' '
            specificWorde = specificWorde.strip()     



            data=[company_name,company_location,date,dateNum,jop_tital,jop_type,jop_level,min_exp,max_exp,related_jop_titel,strongKeyWord,specificWorde,description]          

            with open('Wuzzuf WebScraper Dataset.csv', 'a+', newline='', encoding='UTF8') as f:
                writer = csv.writer(f)
                writer.writerow(data)

            i +=1
            print('compleat row number '+ str(i))

        except:        
              print('error')
        
```

# all page

```python
for j in range (1,15):

    URL = 'https://wuzzuf.net/search/jobs/?a=navbl%7Cspbl&q=data%20science&start={}'.format(j)
    client = urlopen(URL)
    html = client.read()
    client.close()
    soup = BeautifulSoup(html, 'html.parser')
    containers = soup.find_all('div',{'class':'css-1gatmva e1v1l3u10'})
    scraping(containers)
    
    print('compleat page number '+ str(j))
    
    
print('end scraping')
```
